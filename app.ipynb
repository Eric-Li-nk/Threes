```python
import json
from life_cycle.play import *
from ui.user_entries import *

def threes():
    """ Permet d'enchainer les parties au jeu Threes, de reprendre une partie 
        sauvegardée et de sauvegarder une partie en cours
    """
    saisie = get_user_menu(None)
    while saisie != "Q":
        
        if saisie == "N":
            partie = cycle_play(None)
        elif saisie == "L":
            partie = cycle_play(restore_game())
        elif saisie == "S":
            save_game(partie)
        elif saisie == "C":
            partie = cycle_play(partie)
            
        if partie == True:
            saisie = get_user_menu(None)
        else:
            saisie = get_user_menu(partie)
            
```

```python
threes()
```
